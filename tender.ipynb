```python
class Specifications(BaseModel):
    """Information about an equipment."""
    equipment_type: str = Field(description="Type of equipment. Here are some possible labels: 'погрузчик', 'автопогрузчик '")
    purpose: str = Field(description="Purpose of the equipment. Here are some possible labels: 'погрузочно-разгрузочные работы', 'поднятие и перемещение грузов'")
    overall_length_mm: dict = Field(..., description="The minimum and maximum total length of the equipment from the front to the back, measured in millimeters.", example={"min": 1950, "max": 2200})
    overall_width_mm: dict = Field(..., description="The minimum and maximum total width of the equipment at its widest point, measured in millimeters.", example={"min": 1070, "max": 1200})
    max_overall_height_mm: dict = Field(..., description="The minimum and maximum height of the equipment from the ground to the tallest point when the mast is fully lowered, measured in millimeters.", example={"min": 2000, "max": 2120})
    engine_type: str = Field(description="The power source of the equipment, e.g., diesel, electric, gasoline, LPG.")
    engine_power_kw: int = Field(..., description="The power output of the equipment's engine, measured in kilowatts.")
    rated_speed_kmh: int = Field(..., description="The maximum speed at which the equipment can safely operate, measured in kilometers per hour.")
    fuel_consumption_kgh: float = Field(..., description="The amount of fuel the equipment consumes per hour of operation, measured in kilograms per hour.")
    load_capacity_kg: dict = Field(..., description="The minimum and maximum weight the equipment can lift, measured in kilograms.", example={"min": 1500, "max": 1600})
    mast_type: str = Field(..., description="The design of the equipment's lifting mechanism, e.g., simplex, duplex, triplex, quadruplex.")
    mast_height_lowered_mm: int = Field(..., description="The height of the mast when it is fully lowered, measured in millimeters.")
    mast_height_extended_mm: int = Field(..., description="The height of the mast when it is fully extended, measured in millimeters.")
    lift_height_mm: int = Field(..., description="The maximum height to which the forks can raise a load, measured in millimeters.")
    free_lift_mm: int = Field(..., description="Height the equipment can lift its load without extending its mast, measured in millimeters.")
    tilt_angles_forward_deg: dict = Field(..., description="The minimum and maximum degrees the mast can tilt forward.", example={"min": 5, "max": 8})
    tilt_angles_backward_deg: dict = Field(..., description="The minimum and maximum degrees the mast can tilt backward.", example={"min": 7, "max": 12})
    fork_dimensions_mm: dict = Field(..., description="The dimensions of the equipment's forks, represented as a dictionary with keys 'length' (l), 'width' (w), 'height' (h), all measured in millimeters, it can be displayed in the text as 1000x100x40", example={"l": 1000, "w": 100, "h": 40})
    turning_radius_mm: int = Field(..., description="The smallest circular turn an equipment can make, measured in millimeters.")
    tire_type: str = Field(..., description="Type of tires, e.g., cushion (solid) or pneumatic (air-filled).")
    ground_clearance_mm: dict = Field(..., description="The minimum and maximum ground clearance of the equipment, measured in millimeters.", example={"min": 100, "max": 200})
    operator_cabin_type: str = Field(..., description="The type of operator cabin, e.g., open or enclosed.")
    overhead_guard: bool = Field(..., description="Indicates if there is a metal structure on top of the cabin to protect the operator from falling objects.")
    travel_speed_kmph: int = Field(..., description="Maximum speed at which a forklift can move, measured in kilometers per hour.")
    battery_type: str = Field(..., description="The type of battery used in the equipment, e.g., lead-acid, lithium-ion.")
    battery_capacity_ah: int = Field(..., description="The capacity of the equipment's battery, measured in ampere-hours (Ah). This indicates how much charge the battery can store and thus its potential runtime.")
    headlights_low_beam: bool = Field(..., description="is equipped with low beam headlights.")
    headlights_high_beam: bool = Field(..., description="is equipped with high beam headlights.")
    fog_lights: bool = Field(..., description="whether the equipment has fog lights.")
    turn_signals: bool = Field(..., description="is equipped with turn signals.")
    brake_lights: bool = Field(..., description="is equipped with brake lights.")
    reversing_light: bool = Field(..., description="is equipped with a reversing light, which turns on when the equipment is moving backwards.")
    flashing_beacon: bool = Field(..., description="Indicates whether the equipment has a flashing beacon.")
    grip_opening_mm: int = Field(..., description="Max opening between grips in mm.")
    grip_width_mm: int = Field(..., description="Width of grips in mm.")
    pressure_plates: bool = Field(..., description="Presence of pressure plates for even force distribution.", example="цельные")
    contact_plates_coating: str = Field(..., description="Coating type on contact plates for grip and load protection.", example="полиуретановое")
    documents: List[str] = Field(description="List of documents or manuals or warranties required")
    attachments: str = Field(..., description="Additional equipment for specific tasks, e.g., side shifters, fork positioners.")
    number_of_units_for_delivery: int = Field(..., description="The total number of units to be delivered.")
    delivery_location: str = Field(..., description="The address or location where the units will be delivered.")
```

```python
def load_and_extract_content_from_xlsx(file_path):
    loader = UnstructuredExcelLoader(file_path)
    documents = loader.load()

    all_content = ""
    for document in documents:
        all_content += document.page_content

    return all_content.strip()

def extract_metadata_from_pdf_output_parser(file_path):
    pages = load_and_extract_content_from_xlsx(file_path)

    parser = JsonOutputParser(pydantic_object=Specifications)

    prompt = PromptTemplate(
        template="""
        <|system|>
        Extract JSON response as specified in the format instructions. Only extract the properties mentioned:
        {format_instructions}

        </s>
        <|user|>
        Extract entities from the following text in Russian: {context}
        </s>
        <|assistant|>
        """,
        input_variables=["context","format_instructions"]
    )

    chain = prompt | llm | parser

    response = chain.invoke({
        "context": pages,
        "format_instructions": parser.get_format_instructions()
    })

    print(response)
```

```python
def load_pdf(file_path):
    loader = PyPDFLoader(file_path)
    pages = loader.load()
    return pages

def load_xlsl(file_path):
    loader = UnstructuredExcelLoader(file_path)
    docs_xls = loader.load()
    return docs_xls

def load_document(file_path):
    # Determine the file extension
    if file_path.endswith('.pdf'):
        return load_pdf(file_path)
    elif file_path.endswith('.xlsx'):
        return load_xlsl(file_path)
    else:
        raise ValueError("Unsupported file format")

def extract_metadata_from_pdf_output_parser(file_path):
    pages = load_document(file_path)

    parser = JsonOutputParser(pydantic_object=Specifications)

    prompt = PromptTemplate(
        template="""
        <|system|>
        Extract the relevant entities with their properties from the 'context' following 'format instructions'.\
        If an entity is not present and is not required in the instructions, do not include it in the output. Wrap the output in `json`.\
        Context: {context}
        </s>
        <|user|>
        Format instructions: {format_instructions}
        </s>
        <|assistant|>
        """,
        input_variables=["context","format_instructions"]
    )

    chain = prompt | llm | parser

    response = chain.invoke({
        "context": pages,
        "format_instructions": parser.get_format_instructions()
    })

    print(response)
```
